```python
# Splitting Legitimate and Fraudulent Transactions
legit = data[data.Class == 0]
fraud = data[data.Class == 1]

print(f"Legit Transactions: {legit.shape}")
print(f"Fraudulent Transactions: {fraud.shape}")
```

```python
# Balancing the dataset
legit_sample = legit.sample(n=len(fraud))  # Ensures equal fraud and non-fraud samples
new_dataset = pd.concat([legit_sample, fraud])
```

```python
# Shuffle the dataset to mix fraud and non-fraud samples
new_dataset = new_dataset.sample(frac=1).reset_index(drop=True)
```

```python
# Display the new dataset
print(new_dataset.head())
```

```python
# Distribution of Class in the new dataset
print(new_dataset['Class'].value_counts())
```

```python
# Grouping by Class
print(new_dataset.groupby('Class').mean())
```

```python
# Splitting Features (X) and Target (Y)
X = new_dataset.drop(columns='Class', axis=1)
Y = new_dataset['Class']
```

```python
# Splitting into Train and Test datasets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42, stratify=Y)
```

```python
# Checking shapes of data splits
print(f"Shape of X: {X.shape}")
print(f"Shape of X_train: {X_train.shape}")
print(f"Shape of X_test: {X_test.shape}")
```

```python
# Logistic Regression Model
model = LogisticRegression()
```

```python
# Training the Model
model.fit(X_train, Y_train)
```

```python
# Predictions on Training Data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
print(f"Training Data Accuracy: {training_data_accuracy}")
```

```python
# Predictions on Test Data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
print(f"Test Data Accuracy: {test_data_accuracy}")
```
